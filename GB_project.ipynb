```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
raw = pd.read_csv('data/human/Human.CPM.txt', sep = " ")
```

```python
lst = raw.columns.str.rsplit(".",1)
lst2 = [item[0] for item in lst]
raw.columns = lst2
raw.head()
```

```python
med_human = raw.groupby(raw.columns,axis=1,sort=False).median()
med_human.head()
```

```python
#meds.to_csv(r'data/human/Human_CMP_med.csv', index = True)
```

```python
mouse = pd.read_csv('data/mouse/Mouse.CPM.txt', sep = " ")
lst = mouse.columns.str.rsplit(".",1)
mouse.columns = [item[0] for item in lst]
med_mouse = mouse.groupby(mouse.columns,axis=1,sort=False).median()

rat = pd.read_csv('data/rat/Rat.CPM.txt', sep = " ")
lst = rat.columns.str.rsplit(".",1)
rat.columns = [item[0] for item in lst]
med_rat = rat.groupby(rat.columns,axis=1,sort=False).median()

rabbit = pd.read_csv('data/rabbit/Rabbit.CPM.txt', sep = " ")
lst = rabbit.columns.str.rsplit(".",1)
rabbit.columns = [item[0] for item in lst]
med_rabbit = rabbit.groupby(rabbit.columns,axis=1,sort=False).median()

opossum = pd.read_csv('data/opossum/Opossum.CPM.txt', sep = " ")
lst = opossum.columns.str.rsplit(".",1)
opossum.columns = [item[0] for item in lst]
med_opossum = opossum.groupby(opossum.columns,axis=1,sort=False).median()

rhesus = pd.read_csv('data/macaque/Rhesus.CPM.txt', sep = " ")
lst = rhesus.columns.str.rsplit(".",1)
rhesus.columns = [item[0] for item in lst]
med_rhesus = rhesus.groupby(rhesus.columns,axis=1,sort=False).median()

chicken = pd.read_csv('data/chicken/Chicken.CPM.txt', sep = " ")
lst = chicken.columns.str.rsplit(".",1)
chicken.columns = [item[0] for item in lst]
med_chicken = chicken.groupby(chicken.columns,axis=1,sort=False).median()
```

```python
brain = pd.read_csv('data/Brain_genes.csv')
cer = pd.read_csv('data/cerebellum_genes.csv')
heart = pd.read_csv('data/heart_genes.csv')
liver = pd.read_csv('data/liver_genes.csv')
testis = pd.read_csv('data/testis_genes.csv')
brain.head()
```

```python
brain = brain.rename(columns=brain.iloc[1])
brain = brain.drop(brain.index[:2])
cer = cer.rename(columns=cer.iloc[1])
cer = cer.drop(cer.index[:2])
heart = heart.rename(columns=heart.iloc[1])
heart = heart.drop(heart.index[:2])
liver = liver.rename(columns=liver.iloc[1])
liver = liver.drop(liver.index[:2])
testis = testis.rename(columns=testis.iloc[1])
testis = testis.drop(testis.index[:2])
brain.head()
```

```python
genes = pd.concat([brain, cer, heart, liver, testis])
genes = genes.drop(['comparison', 'prob_same', 'on_branch'], axis=1)
genes.head()
```

```python
genes.nunique()
```

```python
cons_genes = genes[genes.classification == 'Conserved']
```

```python
cons_genes.nunique()
```

```python
cons_genes = cons_genes.drop_duplicates()
cons_genes = cons_genes.drop(['classification'], axis=1)
```

```python
merge = pd.merge(cons_genes, med_human, how='left', left_on='Human_ID', right_index=True, sort=True, suffixes=('', '_hum'))
merge = pd.merge(merge, med_mouse, how='left', left_on='Mouse_ID', right_index=True, sort=True, suffixes=('', '_mouse'))
merge = pd.merge(merge, med_rat, how='left', left_on='Rat_ID', right_index=True, sort=True, suffixes=('', '_rat'))
merge = pd.merge(merge, med_rabbit, how='left', left_on='Rabbit_ID', right_index=True, sort=True, suffixes=('', '_rabbit'))
merge = pd.merge(merge, med_opossum, how='left', left_on='Opossum_ID', right_index=True, sort=True, suffixes=('', '_opo'))
merge.head()
```

```python
final = merge.set_index('Human_ID')
final = final.drop(['Mouse_ID', 'Rat_ID', 'Rabbit_ID', 'Opossum_ID'], axis=1)
final.head()
```
